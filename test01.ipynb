```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scanpy as sc
import anndata as ad
```

```python
dirpath = 'test_data\sce.h5ad'

# load the data
adata = sc.read_h5ad(dirpath)

```

```python
expression_matrix = adata.X.toarray() if hasattr(adata.X, 'toarray') else adata.X

df = pd.DataFrame(expression_matrix, index=adata.obs.index, columns=adata.var.index)

# print(df)
df
```

```python
training = pd.read_csv('train/training.csv', names=['UMI', 'Binary'])

df.reset_index(inplace=True)
df.rename(columns={'index': 'UMI'}, inplace=True)
```

```python
mask = df['UMI'].isin(training['UMI'])
training_match = df[mask].copy()

training_match = training_match.reset_index().drop('index', axis=1)
training_match['Binary'] = training['Binary'].reset_index(drop=True)
training_match
```

## Feature selection

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
X = training_match.drop(columns=["UMI", "Binary"]).values
y = training_match["Binary"].values

# 1. feature selection
# using f_classif to select features
selector = SelectKBest(f_classif, k=910)  # change k to select different number of features
X_selected = selector.fit_transform(X, y)


X_train_selected, X_test_selected, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)


clf = LogisticRegression(max_iter=1000, random_state=42)
clf.fit(X_train_selected, y_train)


y_pred_proba = clf.predict_proba(X_test_selected)[:, 1]
loss = log_loss(y_test, y_pred_proba)

print(f"Cross-entropy loss: {loss:.4f}")
```

```python
mask = -df['UMI'].isin(training['UMI'])

testing = df[mask].copy()

testing
```

```python
# feature selection for testing data
X_testing = testing.drop(columns=["UMI"]).values
X_testing_selected = selector.transform(X_testing)  # 注意我们只使用 transform 而不是 fit_transform，因为我们要使用先前的特征选择结果

# 使用模型进行预测
y_testing_pred_proba = clf.predict_proba(X_testing_selected)[:, 1]

y_testing_pred_proba
```

```python
len(y_testing_pred_proba)
```

```python
threshold = 0.5
y_testing_pred_label = (y_testing_pred_proba >= threshold).astype(int)
```

```python
y_testing_pred_label
```

```python
predicted_labels_series = pd.Series(y_testing_pred_label, name='Predicted_Label')
result = pd.concat([testing['UMI'].reset_index(drop=True), predicted_labels_series], axis=1)
```
